# BOOTELO RANKER

```python
# python bootelo.py <ranker_year> <ranker_month> <ranker_day> <config_file>
# Rodar via linha de comando
# Exportar para JSON / CSV / HTML

ranker_year, ranker_month, ranker_day = 2022, 12, 31

game_data_filename = "campeonato-brasileiro-full.csv"
starter_elo = 1500
beta = 250
k_update = 32
ranker_name = "EloRanker"
my_team = "flamengo"
```

```python
import os
from datetime import datetime
from bootelo.ranker.elo_ranker import EloRanker
from bootelo.ranker.trueskill_ranker import TrueskillRanker
from scipy import stats
```

```python
app_folder = os.getcwd()
root_folder = os.path.dirname(app_folder)
data_folder = os.path.join(root_folder, "datasets", "brasileirao")
game_data_filepath = os.path.join(data_folder, game_data_filename)
game_data_filepath
```

```python
ranker = EloRanker(game_data_filepath, beta, starter_elo, k_update=k_update)
```

```python
team_list = ranker.create_ranking(datetime(ranker_year, ranker_month, ranker_day))
```

```python
df_games = ranker.games.to_df()
df_games['year'] = df_games['hour'].dt.year
year_grades = df_games.groupby(["name", "year"]).agg({"grade": "last"}).reset_index()
year_grades['name_year'] = year_grades['name'] + '-' + year_grades['year'].astype(str)

year_grades.to_csv("year_grades.csv", sep=";", decimal=',', index=False)
```

```python
year_grades
```

```python
# Read the CSV file containing team names
# teams_df = pd.read_csv("teams.csv")  # Assuming the file is named "teams.csv"
teams_df = year_grades.copy()

# Print the DataFrame to inspect its structure
print("DataFrame from 'teams.csv':")
# Extract unique team names
teams = teams_df["name_year"].unique()

# FILTRAR OS CAMPEOES

teams
```

```python
teams_df
```

```python
import pandas
from itertools import combinations
# Generate all combinations of teams playing each other in each round
matches = list(combinations(teams, 2))

total_matches = []
for match in matches:
    total_matches.append((match[0], match[1]))
    total_matches.append((match[1], match[0]))

df_matches = pandas.DataFrame(total_matches, columns=["home", "away"])
df_matches = df_matches.merge(teams_df[['name_year', 'grade']], left_on="home", right_on="name_year", how="left")
df_matches = df_matches.merge(teams_df[['name_year', 'grade']], left_on="away", right_on="name_year", how="left", suffixes=('_home', '_away'))
# norm(loc=home.grade - away.grade, scale=2**(1/2)*self.beta).cdf(0)
df_matches['home_prob'] = stats.norm(loc=df_matches['grade_home'] - df_matches['grade_away'], scale=2**(1/2)*beta).cdf(0)

df_matches
```

```python
try:
    # generate a df for the matches 
    rounds = []
    
    for i in range(1, 39):  # 38 rounds
        round_matches = [(home, away) if i % 2 == 0 else (away, home) for home, away in matches]
        rounds.extend(round_matches)
    
    # Create DataFrame for matches
    df_matches = pandas.DataFrame(rounds, columns=["Home", "Away"])
    
    # Export to CSV
    df_matches.to_csv("38_round_matches.csv", index=False)
    
    print("Match table generated and exported successfully.")
except Exception as e:
    print("An error occurred while generating match table:", e)
```
