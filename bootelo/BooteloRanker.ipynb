# BOOTELO RANKER

```python
# python bootelo.py <ranker_year> <ranker_month> <ranker_day> <config_file>
# Rodar via linha de comando
# Exportar para JSON / CSV / HTML

ranker_year, ranker_month, ranker_day = 2022, 12, 31

game_data_filename = "campeonato-brasileiro-full.csv"
starter_elo = 1500
beta = 250
k_update = 24
ranker_name = "EloRanker"
my_team = "flamengo"
```

```python
import os
from datetime import datetime
from bootelo.ranker.elo_ranker import EloRanker
from bootelo.ranker.trueskill_ranker import TrueskillRanker
from scipy import stats
```

```python
app_folder = os.getcwd()
root_folder = os.path.dirname(app_folder)
data_folder = os.path.join(root_folder, "datasets", "brasileirao")
game_data_filepath = os.path.join(data_folder, game_data_filename)
game_data_filepath
```

```python
ranker = EloRanker(game_data_filepath, beta, starter_elo, k_update=k_update)
```

```python
team_list = ranker.create_ranking(datetime(ranker_year, ranker_month, ranker_day))
```

```python
print(f"{my_team.upper()} > position: {ranker.teams[my_team].position} grade: {ranker.teams[my_team].grade}")
```

```python
import seaborn
df_games = ranker.games.to_df()
df_games['year'] = df_games['hour'].dt.year
year_grades = df_games.groupby(["name", "year"]).agg({"grade": "last"}).reset_index()

seaborn.lineplot(data=year_grades, x="year", y="grade", hue="name")
```

```python
import pandas
df = pandas.read_csv(game_data_filepath)

bra_draw_prob = (df['vencedor'] == '-').mean()
bra_draw_prob
```

```python
ts_ranker = TrueskillRanker(game_data_filepath, mu=1500.0, sigma=500.0, beta=250.0, tau=5.0, draw_probability=bra_draw_prob)
```

```python
ts_ranking = ts_ranker.create_ranking(datetime(ranker_year, ranker_month, ranker_day))
```
