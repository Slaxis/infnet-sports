# FIFA PLAYERS CLUSTERING

```python
import os
import pandas
from sklearn.preprocessing import StandardScaler
```

```python
csv_filename = 'players_22.csv'
my_athlete = 'neymar jr'
distance_threshold = 0.5
```

```python
# import pandas
working_dir = os.getcwd()
root_folder = os.path.dirname(working_dir)
datasets_folder = os.path.join(root_folder, 'datasets')
brasileirao_folder = os.path.join(datasets_folder, 'fifa')
csv_path = os.path.join(brasileirao_folder, csv_filename)
csv_path
```

```python
raw = pandas.read_csv(csv_path, sep=',', decimal='.', low_memory=False)

for column in raw.columns:
    print(column)
```

```python
my_cols = ['short_name', 'nationality_name', 'club_name', 'player_positions', 'overall', 'potential', 'wage_eur', 'age', 'height_cm', 'weight_kg', 'preferred_foot', 'international_reputation']

df = raw[my_cols].copy()
df = df.rename(columns={'short_name': 'name', 'player_positions': 'position', 'wage_eur': 'wage', 'preferred_foot': 'foot', 'international_reputation': 'reputation'})
df['name'] = df['name'].str.lower() 
df['right_foot'] = df['foot'].apply(lambda x: 1 if x == 'Right' else 0)
df = df.dropna()
df.head(10)
```

```python
cluster_columns = ['age', 'height_cm', 'weight_kg']
df_cluster = df[cluster_columns].copy()
```

```python
# Create an instance of StandardScaler
scaler = StandardScaler()

# Fit the scaler to your data
scaled_data = scaler.fit_transform(df_cluster)
```

```python
from sklearn.cluster import AgglomerativeClustering

# Create an instance of AgglomerativeClustering
clustering = AgglomerativeClustering(distance_threshold=distance_threshold, n_clusters=None)

# Fit the clustering model to your scaled data
cluster_labels = clustering.fit_predict(scaled_data)
```

```python
n_clusters = clustering.n_clusters_
print(f'Number of clusters found: {n_clusters} @ distance_threshold = {distance_threshold}')
```

```python
df['cluster'] = cluster_labels
```

```python
my_athletes_cluster = df[df['name'] == my_athlete]['cluster'].values[0]

df[df['cluster'] == my_athletes_cluster].sort_values(by='overall', ascending=False)
```
